```python
import yfinance as yf
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# Define the stocks and download data
stocks = ['DIS', 'BIIB', 'PGR', 'SWK', 'PEP', 'ZION', 'LUV', 'BBWI', 'TFC', 'CBRE']
start_date = "2007-02-05"
end_date = "2022-07-18"

stock_data = yf.download(stocks, start=start_date, end=end_date)

# Extract the adjusted closing prices
prices = stock_data['Adj Close']

# Calculate daily returns as percentage change
returns = prices.pct_change().dropna()

# Split the data into training and testing sets (70:30)
train_size = int(len(returns) * 0.7)
train_data, test_data = returns[:train_size], returns[train_size:]

# Loop through each stock and perform ARIMA analysis
for stock in stocks:
    print(f"ARIMA analysis for {stock}:")
    
    # Create a dataframe for the specific stock's returns
    stock_returns = returns[stock].dropna()
    
    # Split the data into training and testing sets
    train_size = int(len(stock_returns) * 0.7)
    train_data, test_data = stock_returns[:train_size], stock_returns[train_size:]
    
    # Fit ARIMA model
    model = ARIMA(train_data, order=(5,1,0))
    model_fit = model.fit()
    
    # Make one-step ahead predictions for percentage change
    predictions = model_fit.forecast(steps=len(test_data))
    
    # Plot the results
    plt.figure(figsize=(12, 6))
    plt.plot(test_data.index, test_data.values, label='Actual Percentage Change')
    plt.plot(test_data.index, predictions, color='red', label='ARIMA Predictions')
    plt.title(f'{stock} Stock Percentage Change ARIMA Prediction')
    plt.xlabel('Date')
    plt.ylabel('Percentage Change')
    plt.legend()
    plt.show()
    
    # Calculate Mean Absolute Error (MAE) as a measure of model accuracy
    mae = np.mean(np.abs(predictions - test_data.values))
    print(f"Mean Absolute Error (MAE) for {stock}: {mae:.4f}")
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import itertools
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_log_error

# Define the stocks and download data
stocks = ['DIS', 'BIIB', 'PGR', 'SWK', 'PEP', 'ZION', 'LUV', 'BBWI', 'TFC', 'CBRE']
start_date = "2007-02-05"
end_date = "2022-07-18"

stock_data = yf.download(stocks, start=start_date, end=end_date)

# Extract the adjusted closing prices
prices = stock_data['Adj Close']

# Calculate daily returns as percentage change
returns = prices.pct_change().dropna()

# Split the data into train, validation, and test sets
train_size = int(len(returns) * 0.7)
valid_size = int(len(returns) * 0.2)
test_size = len(returns) - train_size - valid_size

train_data, valid_data, test_data = (
    returns[:train_size],
    returns[train_size:train_size + valid_size],
    returns[train_size + valid_size:],
)

# Define a parameter grid for ARIMA
param_grid = {
    'p': range(5),     # AR parameter
    'd': range(2),     # Differencing parameter
    'q': range(5)      # MA parameter
}

best_rmse = float('inf')
best_order = None


# Grid search for the best ARIMA parameters based on RMSE
for stock in stocks:
    print(f"Grid search for {stock}:")
    
    for params in itertools.product(param_grid['p'], param_grid['d'], param_grid['q']):
        p, d, q = params
        
        # Fit ARIMA model
        model = ARIMA(train_data[stock], order=(p,d,q))
        model_fit = model.fit()
        
        # Make one-step ahead predictions for the validation set
        predictions = model_fit.forecast(steps=len(valid_data[stock]))
        
        # Calculate RMSE and Huber loss for the validation set
        rmse = np.sqrt(mean_squared_error(valid_data[stock], predictions))
        #huber = np.mean(huber_loss(valid_data[stock], predictions))
        predictions_test = model_fit.forecast(steps = len(test_data[stock]))
        test_rmse = np.sqrt(mean_squared_error(test_data[stock], predictions_test))
        
        if rmse < best_rmse:
            best_rmse = rmse
            best_test_rmse = test_rmse
            best_order = (p, d, q)

            
    print(f"Best ARIMA order for {stock}: {best_order}")
    print(f"Validation RMSE: {best_rmse:.4f}")
    print(f"Test RMSE: {best_test_rmse:.4f}")

# # Use the best ARIMA order to fit the final model on the entire training set
# final_model = ARIMA(train_data[stocks[0]], order=best_order)
# final_model_fit = final_model.fit()

# # Make one-step ahead predictions for the test set
# test_predictions = final_model_fit.forecast(steps=len(test_data[stocks[0]]))

# # Calculate Huber loss for the test set
# test_huber_loss = np.mean(huber_loss(test_data[stocks[0]], test_predictions))

# print("Final Model Performance:")
# print(f"Final ARIMA order: {best_order}")
# print(f"Test RMSE: {best_rmse:.4f}")
# #print(f"Test Huber Loss: {test_huber_loss:.4f}")

# # Plot the final model's predictions against the test data
# plt.figure(figsize=(12, 6))
# plt.plot(test_data.index, test_data[stocks[0]], label='Actual Percentage Change')
# plt.plot(test_data.index, test_predictions, color='red', label='Final ARIMA Predictions')
# plt.title(f'Final ARIMA Model Performance')
# plt.xlabel('Date')
# plt.ylabel('Percentage Change')
# plt.legend()
# plt.show()
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from itertools import product
from sklearn.metrics import mean_squared_error
from math import sqrt

# Define the stocks and download data
stocks = ['DIS', 'BIIB', 'PGR', 'SWK', 'PEP', 'ZION', 'LUV', 'BBWI', 'TFC', 'CBRE']
start_date = "2007-02-05"
end_date = "2022-07-18"

stock_data = yf.download(stocks, start=start_date, end=end_date)

# Extract the adjusted closing prices
prices = stock_data['Adj Close']

# Calculate daily returns as percentage change
returns = prices.pct_change().dropna()

# Divide data into train, validation, and test sets (70%, 20%, 10%)
train_size = int(len(returns) * 0.7)
valid_size = int(len(returns) * 0.2)
test_size = len(returns) - train_size - valid_size

train_data, valid_data, test_data = (
    returns[:train_size],
    returns[train_size:train_size + valid_size],
    returns[train_size + valid_size:],
)

# Define a grid of SARIMA parameters
p = d = q = range(2)  # SARIMA hyperparameters
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in product(p, d, q)]  # Seasonal hyperparameters

best_rmse = float('inf')
best_order = None
best_seasonal_order = None

# Loop through each stock and perform SARIMA analysis with grid search
for stock in stocks:
    print(f"SARIMA analysis for {stock}:")
    
    # Create a dataframe for the specific stock's returns
    stock_returns = returns[stock].dropna()
    
    # Split the data into training, validation, and test sets
    train_size = int(len(stock_returns) * 0.7)
    valid_size = int(len(stock_returns) * 0.2)
    train_data, valid_data, test_data = (
        stock_returns[:train_size],
        stock_returns[train_size:train_size + valid_size],
        stock_returns[train_size + valid_size:],
    )
    
    # Grid search for best SARIMA parameters based on RMSE for the validation set
    for param in seasonal_pdq:
        try:
            model = SARIMAX(train_data, order=(1, 1, 1), seasonal_order=param)
            model_fit = model.fit()
            predictions = model_fit.forecast(steps=len(valid_data))
            rmse = sqrt(mean_squared_error(valid_data, predictions))
            
            if rmse < best_rmse:
                best_rmse = rmse
                best_order = (1, 1, 1)
                best_seasonal_order = param
        except:
            continue
    
    # Fit SARIMA model with the best parameters
    model = SARIMAX(train_data, order=best_order, seasonal_order=best_seasonal_order)
    model_fit = model.fit()
    
    # Make one-step ahead predictions for percentage change
    predictions = model_fit.forecast(steps=len(test_data))
    
    # Plot the results
    plt.figure(figsize=(12, 6))
    plt.plot(test_data.index, test_data.values, label='Actual Percentage Change')
    plt.plot(test_data.index, predictions, color='red', label='SARIMA Predictions')
    plt.title(f'{stock} Stock Percentage Change SARIMA Prediction')
    plt.xlabel('Date')
    plt.ylabel('Percentage Change')
    plt.legend()
    plt.show()
    
    # Calculate RMSE as a measure of model accuracy for the test set
    rmse = sqrt(mean_squared_error(test_data, predictions))
    print(f"RMSE for {stock}: {rmse:.4f}")
    print(f"Best SARIMA order for {stock}: {best_order}")
    print(f"Best Seasonal Order for {stock}: {best_seasonal_order}")
```
